```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from IPython.display import display, Markdown

# ==========================================
# 1. GENERATE ADVANCED BUSINESS DATASET
# ==========================================
data = {
    'Marketing_Channel': ['Organic Search', 'Paid Social (Meta)', 'Paid Search (Google)', 'Email Marketing'],
    'Website_Visitors': [18000, 32000, 15000, 5000],
    'Generated_Leads': [3600, 2500, 1800, 2200],
    'Paying_Customers': [720, 200, 270, 880],
    'Marketing_Spend_USD': [1200, 18000, 10000, 800],
    'Revenue_Generated_USD': [72000, 20000, 27000, 88000]
}

df = pd.DataFrame(data)

# ==========================================
# 2. CALCULATE ADVANCED KPI METRICS
# ==========================================
df['Traffic_to_Lead_Rate (%)'] = round((df['Generated_Leads'] / df['Website_Visitors']) * 100, 2)
df['Lead_to_Customer_Rate (%)'] = round((df['Paying_Customers'] / df['Generated_Leads']) * 100, 2)
df['Overall_Conversion (%)'] = round((df['Paying_Customers'] / df['Website_Visitors']) * 100, 2)
df['CAC_USD'] = round(df['Marketing_Spend_USD'] / df['Paying_Customers'], 2)
df['ROAS'] = round(df['Revenue_Generated_USD'] / df['Marketing_Spend_USD'], 2)

# ==========================================
# 3. BUILD THE INTERACTIVE DASHBOARD
# ==========================================

# Chart 0: Interactive Data Table
headers = [col.replace('_', ' ') for col in df.columns]
fig0 = go.Figure(data=[go.Table(
    header=dict(values=[f"<b>{h}</b>" for h in headers], fill_color='#1F2937', font=dict(color='white', size=12), align='center'),
    cells=dict(values=[df[col] for col in df.columns], fill_color='#F9FAFB', font=dict(color='black', size=12), align='center'))
])
fig0.update_layout(title_text="Executive KPI Dataset", title_x=0.5, margin=dict(t=40, b=10), height=250)
fig0.show()

# Chart 1: The Macro Funnel
total_visitors = df['Website_Visitors'].sum()
total_leads = df['Generated_Leads'].sum()
total_customers = df['Paying_Customers'].sum()

fig1 = go.Figure(go.Funnel(
    y=['Website Visitors', 'Generated Leads', 'Paying Customers'],
    x=[total_visitors, total_leads, total_customers],
    textinfo="value+percent initial",
    marker={"color": ["#1F2937", "#3B82F6", "#10B981"]}
))
fig1.update_layout(title_text="1. Overall Marketing Funnel Drop-off", title_x=0.5, template="plotly_white")
fig1.show()

# Chart 2: Channel Conversion Breakdown
fig2 = go.Figure()
fig2.add_trace(go.Bar(x=df['Marketing_Channel'], y=df['Website_Visitors'], name='Visitors', marker_color='#9CA3AF'))
fig2.add_trace(go.Bar(x=df['Marketing_Channel'], y=df['Generated_Leads'], name='Leads', marker_color='#F59E0B'))
fig2.add_trace(go.Bar(x=df['Marketing_Channel'], y=df['Paying_Customers'], name='Customers', marker_color='#10B981'))
fig2.update_layout(title_text="2. Funnel Performance by Marketing Channel", title_x=0.5, barmode='group', template="plotly_white")
fig2.show()

# Chart 3: Business Impact (CAC by Channel)
fig3 = px.bar(
    df,
    x='Marketing_Channel',
    y='CAC_USD',
    color='ROAS',
    text='CAC_USD',
    color_continuous_scale='RdYlGn',
    title="3. Business Impact: Customer Acquisition Cost (CAC) & ROAS"
)
fig3.update_traces(texttemplate='$%{text}', textposition='outside')
fig3.update_layout(title_x=0.5, yaxis_title="Cost to Acquire One Customer ($)", template="plotly_white")
fig3.show()

# ==========================================
# 4. PRINT EXECUTIVE REPORT
# ==========================================
report = """
---
## 🎯 Executive Business Insights & Recommendations

**1. Where are users dropping off in the funnel?**
The most significant drop-off occurs at the very top of the funnel between the **Website Visitors** and **Generated Leads** stages. Out of 70,000 total visitors, only 10,100 converted into leads, representing a massive **85.5% abandonment rate** before we even capture their contact information.

**2. Which channels bring high-quality leads?**
**Email Marketing** brings in the highest-quality leads by a landslide. Even though it generated the lowest amount of initial traffic, **40%** of the leads from this channel converted into paying customers.

**3. How can conversion rates be improved?**
Conversion rates can be drastically improved by reallocating the marketing budget. Currently, $18,000 is being spent on "Paid Social", which has the worst overall conversion rate and an unsustainable Customer Acquisition Cost (CAC) of $90. Shifting that budget to aggressively grow the "Email Marketing" subscriber list will immediately boost the overall conversion rate and lower CAC.

**4. Which stages need optimization?**
The **top-of-funnel landing pages** need immediate optimization. Because 85.5% of visitors leave without becoming a lead, the business needs to A/B test stronger Call-to-Actions (CTAs) and introduce high-value "Lead Magnets" (like free industry reports or templates) to capture emails earlier.
"""
display(Markdown(report))
```
